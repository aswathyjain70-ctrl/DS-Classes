**TASK 7**

```python
population = pd.to_numeric(population['population'], errors='coerce')
population
```

```python
gdp = pd.to_numeric(gdp ['GDP'], errors='coerce')
gdp
```

```python
internet_users = pd.to_numeric(internet_users['internet_users'], errors='coerce')
internet_users
```

```python
literacy = pd.to_numeric(literacy['literacy_rate'], errors='coerce')
literacy
```

**TASK 8**

```python
df = population.merge(gdp, on='country', how='inner')
df = df.merge(internet_users, on='country', how='inner')
df = df.merge(literacy, on='country', how='inner')

df.head()
```

**TASK 9**

```python
df['Internet Penetration (%)'] = ((df['internet_users'] / df['population']) * 100).round(2)
df['GDP Per Capita'] = (df['GDP'] / df['population']).round(2)
df.head()
```

**TASK** 10

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('world_data.db')
df.to_sql('country_statistics', conn, if_exists='replace', index=False)
query = "SELECT * FROM country_statistics LIMIT 5;"
verified_df = pd.read_sql(query, conn)
conn.close()
verified_df
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('world_data.db')
```

```python
query= """
SELECT country, GDP
FROM country_statistics
ORDER BY GDP DESC
LIMIT 5;
"""
pd.read_sql_query(query, conn)

```

```python
query_2 = """
SELECT AVG(literacy_rate) AS average_literacy_rate
FROM country_statistics;
"""
pd.read_sql_query(query_2, conn)
```

```python
query_3 = """
SELECT country, literacy_rate
FROM country_statistics
WHERE literacy_rate > 90;
"""
pd.read_sql_query(query_3, conn)
```

**Q4**

```python
import numpy as np
random_matrix = np.random.rand(5,4)
print(random_matrix)

```

```python
import numpy as np
row_sums = np.sum(random_matrix, axis=1)
max_row_idx = np.argmax(row_sums)
max_sum_row = random_matrix[max_row_idx]
print(row_sums)
print(max_row_idx)
print(max_sum_row)
```

```python
import numpy as np
row_min = np.min(random_matrix, axis=1).reshape(-1,1)
row_max = np.max(random_matrix, axis=1).reshape(-1,1)
normalized_matrix = (random_matrix - row_min) / (row_max - row_min)
print("normalized Matrix (each row scaled between 0 and 1):")
print(normalized_matrix)
```

**Q8**

```python
revenue_median = data['Revenue (Millions)'].median()
data['Revenue (Millions)'].fillna(revenue_median)

metascore_median = data['Metascore'].median()
data['Metascore'].fillna(metascore_median)

data[['Revenue (Millions)', 'Metascore']].isnull().sum()

```

**Q9**

```python
top_rated_movies = data.nlargest(5, 'Rating')
top_rated_movies[['Movie_Title', 'Rating']]
```

**Q10**

```python
highest_revenue_movies = data.nlargest(5, 'Revenue (Millions)')
highest_revenue_movies[['Movie_Title', 'Revenue (Millions)']]
```

```python
lowest_revenue_movies = data.nsmallest(5, 'Revenue (Millions)')
lowest_revenue_movies[['Movie_Title', 'Revenue (Millions)']]
```

**Q11**

```python
avg_rating_by_director = data.groupby('Director')['Rating'].mean()
avg_rating_by_director.head(10)
```

**Q12**

```python
movies_per_year = data ['Year'].value_counts()
movies_per_year.sort_index()
```

**Q13**

```python
movies = data[(data["Rating"] > 8) & (data["Revenue (Millions)"] > 100)][["Movie_Title","Rating","Revenue (Millions)"]]
movies


```

**Q14**

```python
director_stats = data.groupby('Director').agg({
    'Rating': 'mean',
    'Revenue (Millions)': 'mean'
}).reset_index()

director_stats.columns = ['Director', 'Avg_Rating', 'Avg_Revenue_Millions']

director_stats = director_stats.sort_values(by='Avg_Rating', ascending=False)
director_stats.head(10)
```

**Q15**

```python
year_stats = data.groupby('Year').agg(
    Movie_count = ('Movie_Title' , 'count'),
    Avg_Rating = ('Rating', 'mean')
).reset_index()

year_stats.sort_values(by='Year')
```

**Q16**

```python
data['Profitability'] = data['Revenue (Millions)'] - data['Runtime (Minutes)']
top_10_profitable_movies = data.nlargest(10, 'Profitability')
top_10_profitable_movies[['Movie_Title', 'Profitability']]
```

**Q17**

```python
data['Rating_Category'] = np.where(data['Rating'] < 6, 'Poor',
                          np.where(data['Rating'] < 7.5 , 'Average', 'Excellent'))
data[['Movie_Title', 'Rating', 'Rating_Category']].head(10)
```

**Q18**

```python
data['High_Score'] = data.apply(lambda row: row['Rating'] * 10 + row['Revenue (Millions)'], axis=1)
data[['Movie_Title' , 'High_Score']].head(10)
```

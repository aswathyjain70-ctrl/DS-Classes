```python
class BankAccount:
  accStatus = "Active"
  def __init__(self,owner,accno,balance):
    self.owner = owner
    self.accno = accno
    self.balance = balance

  def deposit(self,amount):
    if amount >0:
      self.balance +=amount
      print(f"Deposited {amount:.2f}. Balance: {self.balance:.2f}")
    else:
      print("Not possible")

  def withdraw(self,amount):
    if amount >0:
      if self.balance >=amount:
        self.balance -= amount
        print(f"withdrew {amount:.2f}. Balance: {self.balance:.2f}")
      else:
        print("Insufficient balance")
    else:
      print("Not possible")

  def balance_inquiry(self):
    print(f"balanace inquiry: {self.balance:.2f}")


  def display(self):
    print(f"Owner : {self.owner} Accno : {self.accno} Balance : {self.balance:.2f}")

cust1 = BankAccount("Ashwin",104490,2200)
cust2 = BankAccount("Rinu",183765,2100)
cust3 = BankAccount("Maya",134440,3000)

cust1.display()
cust1.deposit(500)
cust1.display()
cust2.display()
cust2.balance_inquiry()

cust1.deposit(-100)
cust1.withdraw(7000)
```
